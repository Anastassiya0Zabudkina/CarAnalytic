```python
#Добавить колонку price_tenge, заполнить ее ценой в тенге 
#(для этого берете оригинальную цену машины и умножаете на курс валюты, которая появилась в виде последнего столбца)

exchange_rates = {"dollar": 390, "euro": 455}

# добавление столбца price_tenge
carsDF["price_tenge"] = carsDF.apply(lambda row: row["price"] * exchange_rates[row["currency"]], axis=1)
```

```python
carsDF
```

```python
#Замените колонку с валютой на ее первую букву 
carsDF['currency'] = carsDF['currency'].apply(lambda x: x[0])
```

```python
carsDF['currency'] = carsDF.currency.str[0]
```

```python
#Добавьте новый столбец tax (налог), который равен 1% от цены машины, 
#если она не старше 5 лет, 2% от цены если она от 5 до 10 лет, и 3% если старше 10 лет
import datetime

def tax_func(year, price):
    if datetime.datetime.now().year - year <= 5:
        price *= 0.01
    elif  datetime.datetime.now().year - year > 5 and datetime.datetime.now().year - year <= 10:
        price *= 0.02
    else:
        price *= 0.03
    return price


carsDF['tax'] = carsDF.apply(lambda carsDF: tax_func(carsDF['year'], carsDF['price_tenge']), axis=1)
```
